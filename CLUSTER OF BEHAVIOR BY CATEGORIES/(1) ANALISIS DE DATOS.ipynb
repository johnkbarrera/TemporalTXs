# Analisis de Datos

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import EngFormatter
formatter = EngFormatter(places=0, sep=u"\N{THIN SPACE}")  # U+2009
```

## Cargar Datos

```python
import datetime
import pandas as pd

def leer_data():
    outfile='./SOURCES/consulta.csv'
    data = pd.read_csv(outfile)
    return data

data = leer_data()
print('Done')
```

```python
data.head()
```

```python
temp = data
```

```python
temp['date'] = pd.to_datetime(temp['date'])
temp['dia_semana'] = temp['date'].dt.dayofweek
days = {0:'Mon',1:'Tues',2:'Weds',3:'Thurs',4:'Fri',5:'Sat',6:'Sun'}
temp['dia_semana'] = temp['dia_semana'].apply(lambda x: days[x])
```

```python
days = {0:'1-Madrugada',1:'1-Madrugada',2:'1-Madrugada',3:'1-Madrugada',4:'1-Madrugada',5:'1-Madrugada',
        6:'2-Mañana',7:'2-Mañana',8:'2-Mañana',9:'2-Mañana',10:'2-Mañana',11:'2-Mañana',
        12:'3-Tarde',13:'3-Tarde',14:'3-Tarde',15:'3-Tarde',16:'3-Tarde',17:'3-Tarde',
        18:'4-Noche',19:'4-Noche',20:'4-Noche',21:'4-Noche',22:'4-Noche',23:'4-Noche'}
temp['turno'] = temp['hora'].apply(lambda x: days[x])
```

```python
data =  temp
data.head(3)
```

```python
y = list(data.groupby(['merchant_district']).sum()['amount_sol'])
x = list(data.groupby(['merchant_district']).sum()['amount_sol'] .index)
x[y.index(min(y))]
```

```python
y = list(data.groupby(['merchant_district']).count()['amount_sol'])
x = list(data.groupby(['merchant_district']).count()['amount_sol'] .index)
plt.pie(y, labels=x, autopct='%1.1f%%', shadow=False, startangle=230)
plt.axis('equal') 
for i in range(len(y)):
    print(i,':  ',y[i],'     ',x[i])
```

## Graficos

### Comportamiento de Gastos totales por meses

```python
# Grafico por Meses - Cantidad

nn= data.groupby(['año','mes'], as_index=False, sort=True)['client_id'].count()

fig, ax = plt.subplots()
ax.bar(nn.index, nn['client_id'], align='center')
ax.grid(color='gray', linestyle='--', linewidth=1)

ax.set_xlabel('Año')
ax.set_xticks(nn.index)

ax.set_ylabel('Frecuencia')
ax.set_yticks(nn['client_id'])
ax.set_title('Histograma')
plt.show()
nn
```

```python
# Grafico por Meses - Montos 

nn= data.groupby(['año','mes'], as_index=False, sort=True)['amount_sol'].sum()

fig, ax = plt.subplots()
ax.bar(nn.index, nn['amount_sol'], align='center')
ax.grid(color='gray', linestyle='-', linewidth=1)

ax.set_xlabel('Año')
ax.set_xticks(nn.index)

ax.set_ylabel('Monto Total')
ax.set_yticks(nn['amount_sol'])
ax.set_title('Histograma')
plt.show()
nn
```

## Graficos para cada MCCG

```python
# OBTENIENDO LISTA - MCCG
mccgs = data.groupby(['mccg']).min().index.values
mccgs
```

```python
box = dict(facecolor='#ff9999', pad=4, alpha=0.2)
    
for i in range(len(mccgs)):
#for i in range(1):
    data_temp = data.loc[data['mccg'] == mccgs[i]]
    #data_temp = data
    fig, axes = plt.subplots(2, 4, figsize=(15,12))
    
    
    plt.suptitle('Para el mccg '+str(mccgs[i]), fontsize=16,y=0.989)
    ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8 = axes.flatten()
    plt.subplots_adjust(top=0.9, bottom=0.5, left=0.10, right=0.95, hspace=0.9, wspace=0.55)

    
    ax1.xaxis.set_major_formatter(formatter)
    ax1.yaxis.set_major_formatter(formatter)
    y = data_temp.groupby(['client_id']).count()      # Us por cliente
    y = y["date"].tolist()
    mu = np.mean(y)
    median = np.median(y)
    sigma = np.var(y)
    text = '$\mu=%.2f$\n$\mathrm{median}=%.2f$\n$\sigma=%.2f$' % (mu, median, sigma)
    ax1.hist(y,100, color="#47d147")   #ax1.hist(y,100)
    ax1.set_title('Cantidad de TXs mas Frecuentes')
    ax1.set_xlabel('TXs Sessions', bbox=box)
    ax1.set_ylabel('N° de Clientes', bbox=box)
    ax1.text(0.27, 0.85, text, transform=ax1.transAxes, fontsize=8, verticalalignment='top', bbox=box)

    ax2.xaxis.set_major_formatter(formatter)
    ax2.yaxis.set_major_formatter(formatter)
    y = list(data_temp.groupby(['hora']).sum()['amount_sol'])
    x = range(len(y))
    ax2.plot(x,y,'-',color='#e6e600')
    ax2.plot(x,y,'ob',color='#999900')
    ax2.set_title('Flujo del dinero por Horas')
    ax2.set_xlabel('Horas', bbox=box)
    ax2.set_ylabel('Soles', bbox=box)


    y = data_temp.groupby(['dia_semana']).sum()['amount_sol']
    x = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
    ax3.yaxis.set_major_formatter(formatter)
    ax3.set_xticklabels(x, rotation=45)
    ax3.bar(y.index,y,0.7,color='#ff8080')
    ax3.plot(y.index,y,'-',color='#ff0000')
    ax3.plot(y.index,y,'ob',color='#990000')
    ax3.set_title('Flujo del dinero por Dia')
    ax3.set_xlabel('Horas', bbox=box)
    ax3.set_ylabel('Soles', bbox=box)


    y = list(data_temp.groupby(['turno']).sum()['amount_sol'])
    x = list(data_temp.groupby(['turno']).sum()['amount_sol'] .index)
    ax4.yaxis.set_major_formatter(formatter)
    ax4.set_xticklabels(x, rotation=45)
    ax4.plot(x,y,'-',color='#0052cc')
    ax4.plot(x,y,'ob',color='#003380')
    ax4.set_title('Flujo del dinero por Turno')
    ax4.set_xlabel('Turnos', bbox=box)
    ax4.set_ylabel('Soles', bbox=box)


    ax5.xaxis.set_major_formatter(formatter)
    ax5.yaxis.set_major_formatter(formatter)
    y = data_temp.groupby(['client_id']).sum()
    y2 = y["amount_sol"].tolist()
    y = y.groupby(['amount_sol']).count() 
    mu = np.mean(y2)
    median = np.median(y2)
    sigma = np.var(y2)
    text = '$\mu=%.2f$\n$\mathrm{median}=%.2f$\n$\sigma=%.2f$' % (mu, median, sigma)
    ax5.bar(y.index.tolist(),y["quantity"].tolist(),100,color='#47d147')   #ax1.hist(y,100)
    ax5.set_title('Monto mas Frecuentes en la TXs de los C')
    ax5.set_xlabel('Monto en TXs Sessions', bbox=box)
    ax5.set_ylabel('N° de Clientes', bbox=box)
    ax5.text(0.27, 0.85, text, transform=ax5.transAxes, fontsize=8, verticalalignment='top', bbox=box)


    ax6.xaxis.set_major_formatter(formatter)
    ax6.yaxis.set_major_formatter(formatter)
    y = list(data_temp.groupby(['hora']).count()['quantity'])
    x = range(len(y))
    ax6.plot(x,y,'-',color='#e6e600')
    ax6.plot(x,y,'ob',color='#999900')
    ax6.set_title('N° de TXs por Horas')
    ax6.set_xlabel('Horas', bbox=box)
    ax6.set_ylabel('Cantidad de TXs', bbox=box)


    y = data_temp.groupby(['dia_semana']).count()['quantity']
    x = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
    ax7.yaxis.set_major_formatter(formatter)
    ax7.set_xticklabels(x, rotation=45)
    ax7.bar(y.index,y,color='#ff8080')
    ax7.plot(y.index,y,'-',color='#ff0000')
    ax7.plot(y.index,y,'ob',color='#990000')
    ax7.set_title('N° de TXs por Dia')
    ax7.set_xlabel('Horas', bbox=box)
    ax7.set_ylabel('Cantidad de TXs', bbox=box)


    y = list(data_temp.groupby(['turno']).count()['quantity'])
    x = list(data_temp.groupby(['turno']).count()['quantity'] .index)
    ax8.yaxis.set_major_formatter(formatter)
    ax8.set_xticklabels(x, rotation=45)
    ax8.plot(x,y,'-',color='#0052cc')
    ax8.plot(x,y,'ob',color='#003380')
    ax8.set_title('N° de TXs por Turno (U)')
    ax8.set_xlabel('Turnos', bbox=box)
    ax8.set_ylabel('Cantidad de TXs', bbox=box)

plt.savefig("plot.png")
```
